```python
engine = create_engine('sqlite:///Database1.db')
#then we will create the sqllite connection by creating an engine and downloading the required Apple stock price data
df = yf.download('AAPL', start = '2022-01-01', end = '2023-04-30')
#we will use the df to save the file to a pytho dataframe
```

```python
def add_data_to_database(data, table):# l created this function to make the whole process simpler and be excutyed after a function call
    max_date = pd.read_sql(f'SELECT MAX(Date) FROM {table}', engine).values[0][0]
    print(max_date)#we need to print the last date of the database

    new_rows = data[data.index > max_date]
    new_rows.to_sql(table, engine, if_exists='append')
    print(str(len(new_rows))+ ' new rows imported to Database')# this will print the number of new rows added to the database
```

```python
add_data_to_database(df, 'AAPL')
#then we try to call the function after new data has been added
```
